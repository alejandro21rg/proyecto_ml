# 1. Carga del dataset Elo

Se carga el dataset histórico de Elo Ratings que será utilizado posteriormente para enriquecer los partidos con información de fuerza relativa de las selecciones.

```python
import pandas as pd
import numpy as np

elo = pd.read_csv("data/eloratings.csv")
```

# 6. Conversión de variables temporales

```python
elo["date"] = pd.to_datetime(
    elo["date"],
    format="mixed"
)
```

# 7. Análisis temporal

```python
elo["date"].min()
```

```python
elo["date"].max()
```

# 8. Filtrado temporal definitivo

Se conservan únicamente los registros comprendidos entre los años 2000 y 2025 para mantener coherencia temporal con el resto de datasets utilizados en el proyecto.

```python
elo = elo[
    (elo["date"] >= "2000-01-01") &
    (elo["date"] <= "2025-12-31")
]
```

```python
elo["date"].min()
```

```python
elo["date"].max()
```

```python
elo.shape
```

# 9. Selección de variables finales

```python
elo_clean = elo[
    [
        "date",
        "team",
        "rating"
    ]
]
```
