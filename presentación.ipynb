# Predicción y Simulación del Mundial de Fútbol 2026 mediante Machine Learning

## Objetivo 
Desarrollar un sistema de Machine Learning capaz de predecir resultados de partidos internacionales y simular el Mundial 2026 para estimar probabilidades de clasificación y campeonato.

### Desarrollar un modelo capaz de:
1. Predecir la diferencia de goles entre dos selecciones.
2. Simular el Mundial 2026 completo.
3. Calcular probabilidades de:
   - Pasar de grupo
   - Llegar a cuartos
   - Llegar a semifinales
   - Llegar a la final
   - Ganar el Mundial
