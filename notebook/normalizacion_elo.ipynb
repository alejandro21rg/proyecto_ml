# 1. Carga

```python
import pandas as pd

results = pd.read_csv(
    "../data/processed/results_normalized.csv"
)

elo = pd.read_csv(
    "../data/processed/elo_final.csv"
)
```

# Paso 2. Limpiar caracteres extraños del Elo

```python
elo["team"] = (
    elo["team"]
    .str.replace("\xa0", " ", regex=False)
    .str.strip()
)
```

# 3. Obtener selecciones únicas

```python
teams_results = set(
    results["home_team"]
).union(
    set(results["away_team"])
)
```

```python
teams_elo = set(
    elo["team"]
)
```

```python
print(
    f"Selecciones Results: {len(teams_results)}"
)

print(
    f"Selecciones Elo: {len(teams_elo)}"
)
```

```python
results_not_elo = sorted(
    teams_results - teams_elo
)

len(results_not_elo)
```

```python
results_not_elo
```

```python
elo_not_results = sorted(
    teams_elo - teams_results
)

len(elo_not_results)
```

```python
elo_not_results
```

```python
name_mapping_elo = {
    "American Samoa": "Samoa",
    "Brunei Darussalam": "Brunei",
    "Cabo Verde": "Cape Verde",
    "China PR": "China",
    "Chinese Taipei": "Taiwan",
    "Congo DR": "Democratic Republic of Congo",
    "Côte d'Ivoire": "Ivory Coast",
    "Hong Kong, China": "Hong Kong",
    "IR Iran": "Iran",
    "Korea DPR": "North Korea",
    "Korea Republic": "South Korea",
    "Kyrgyz Republic": "Kyrgyzstan",
    "Macau": "Macao",
    "Republic of Ireland": "Ireland",
    "São Tomé and Príncipe": "Sao Tome and Principe",
    "Timor-Leste": "East Timor",
    "Türkiye": "Turkey",
    "USA": "United States"
}
```

```python
results["home_team"] = (
    results["home_team"]
    .replace(name_mapping_elo)
)

results["away_team"] = (
    results["away_team"]
    .replace(name_mapping_elo)
)
```

```python
teams_results = set(
    results["home_team"]
).union(
    set(results["away_team"])
)

teams_elo = set(
    elo["team"]
)
```

```python
results_not_elo = sorted(
    teams_results - teams_elo
)

len(results_not_elo)
```

```python
results_not_elo
```
