# 3. Detección de valores nulos

Se analiza la presencia de valores faltantes para determinar si es necesario realizar imputaciones o eliminar registros.

```python
fifa.isnull().sum()
```

```python
fifa = fifa.dropna(
    subset=["rank"]
)
```

# 4. Análisis de Duplicados


```python
fifa.duplicated().sum()
```

# 5. Eliminación de columnas irrelevantes

El dataset contiene algunas variables que no aportarán información relevante al modelo predictivo o que generan información redundante.

Se eliminan aquellas columnas que no serán utilizadas durante la fase de modelado.

```python
fifa = fifa.drop(
    columns=["Unnamed: 0"]
)
```

# 5. Conversión de Variables Temporales

La variable rank_date se convierte al formato datetime para facilitar futuras operaciones temporales y la integración con los resultados históricos de los partidos.

```python
fifa["rank_date"] = pd.to_datetime(
    fifa["rank_date"]
)
```

# 9. Dataset final FIFA


```python
fifa = fifa[
    [
        "rank_date",
        "country_full",
        "rank",
        "total_points"
    ]
]
```

```python
fifa = fifa[
    (fifa["rank_date"] >= "2000-01-01") &
    (fifa["rank_date"] <= "2025-12-31")
]
```
