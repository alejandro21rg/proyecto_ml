Me quedo con las columnas necesarias

```python
fifa = ranking_fifa[
    [
        "country_full",
        "rank_date",
        "rank",
        "total_points"
    ]
].copy()
```

ordenar fechas

```python
fifa = fifa.sort_values("rank_date")

df_ml = df_matches.sort_values("date")
```

```python
home_fifa = fifa.rename(
    columns={
        "country_full": "home_team",
        "rank": "fifa_rank_home",
        "total_points": "fifa_points_home",
        "rank_date": "date"
    }
)
```

```python
df_ml = pd.merge_asof(
    df_ml.sort_values("date"),
    home_fifa.sort_values("date"),
    on="date",
    by="home_team",
    direction="backward"
)
```

```python
df_ml[
    [
        "home_team",
        "date",
        "fifa_rank_home",
        "fifa_points_home"
    ]
].head()
```

```python
away_fifa = fifa.rename(
    columns={
        "country_full": "away_team",
        "rank": "fifa_rank_away",
        "total_points": "fifa_points_away",
        "rank_date": "date"
    }
)
```

```python
df_ml = pd.merge_asof(
    df_ml.sort_values("date"),
    away_fifa.sort_values("date"),
    on="date",
    by="away_team",
    direction="backward"
)
```

```python
df_ml[
    [
        "home_team",
        "away_team",
        "fifa_rank_home",
        "fifa_rank_away",
        "fifa_points_home",
        "fifa_points_away"
    ]
].head()
```

```python
df_ml["rank_diff"] = (
    df_ml["fifa_rank_home"]
    - df_ml["fifa_rank_away"]
)

df_ml["points_diff"] = (
    df_ml["fifa_points_home"]
    - df_ml["fifa_points_away"]
)
```
