# 1. Carga de datasets limpios
Se cargan los datasets previamente limpiados para analizar posibles diferencias en la nomenclatura de las selecciones nacionales.

```python
import pandas as pd
import numpy as np

results = pd.read_csv("data/results_final.csv")
fifa = pd.read_csv("data/fifa_final.csv")
```

# 2. Conversión de variables temporales

```python
results["date"] = pd.to_datetime(
    results["date"]
)

fifa["rank_date"] = pd.to_datetime(
    fifa["rank_date"]
)
```

# 3. Extracción de selecciones únicas

Se obtienen los nombres únicos de las selecciones presentes en ambos datasets para detectar posibles diferencias de nomenclatura.


```python
teams_results = set(
    results["home_team"]
).union(
    set(results["away_team"])
)

teams_fifa = set(
    fifa["country_full"]
)
```

```python
print(f"Selecciones Results: {len(teams_results)}")
print(f"Selecciones FIFA: {len(teams_fifa)}")
```

# 5. Comparación de nomenclaturas

Se identifican las selecciones que aparecen en un dataset pero no en el otro con el objetivo de construir posteriormente un diccionario de equivalencias.

```python
results_not_fifa = sorted(
    teams_results - teams_fifa
)

results_not_fifa
```

```python
fifa_not_results = sorted(
    teams_fifa - teams_results
)

fifa_not_results
```

```python
valid_teams = teams_fifa
```

```python
results_fifa = results[
    (results["home_team"].isin(valid_teams)) &
    (results["away_team"].isin(valid_teams))
]
```

```python
results.shape
```

```python
results_fifa.shape
```

```python
teams_results = set(
    results_fifa["home_team"]
).union(
    set(results_fifa["away_team"])
)

teams_fifa = set(
    fifa["country_full"]
)
```

```python
results_not_fifa = sorted(
    teams_results - teams_fifa
)

results_not_fifa
```

```python
name_mapping = {
    "United States": "USA",
    "South Korea": "Korea Republic",
    "North Korea": "Korea DPR",
    "Iran": "IR Iran",
    "Turkey": "Türkiye",
    "Ivory Coast": "Côte d'Ivoire",
    "DR Congo": "Congo DR",
    "Cape Verde": "Cabo Verde",
    "Czech Republic": "Czechia",
    "Hong Kong": "Hong Kong, China",
    "Kyrgyzstan": "Kyrgyz Republic",
    "Taiwan": "Chinese Taipei",
    "Brunei": "Brunei Darussalam"
}
```

```python
results["home_team"] = (
    results["home_team"]
    .replace(name_mapping)
)

results["away_team"] = (
    results["away_team"]
    .replace(name_mapping)
)
```

```python
results_fifa = results[
    (results["home_team"].isin(teams_fifa)) &
    (results["away_team"].isin(teams_fifa))
]

results.shape, results_fifa.shape
```

```python
results_not_fifa = sorted(
    teams_results - teams_fifa
)

len(results_not_fifa)
```

```python
for team in results_not_fifa:
    print(team)
```

```python
teams_results_fifa = set(
    results_fifa["home_team"]
).union(
    set(results_fifa["away_team"])
)
```

```python
results_not_fifa = sorted(
    teams_results_fifa - teams_fifa
)

len(results_not_fifa)
```

```python
excluded = results[
    ~(
        results["home_team"].isin(teams_fifa) &
        results["away_team"].isin(teams_fifa)
    )
]
```
