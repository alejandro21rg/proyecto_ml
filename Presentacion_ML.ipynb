# Predicción y simulación del Mundial 2026 mediante Machine Learning


Objetivo:
Desarrollar un modelo de Machine Learning capaz de predecir resultados de partidos internacionales y utilizarlo para simular el Mundial 2026 mediante técnicas de Monte Carlo.

# 1. Objetivo del proyecto

El objetivo principal consiste en estimar qué selecciones tienen mayor probabilidad de ganar el Mundial 2026.

Para ello se han desarrollado:

- Un modelo predictivo de resultados.
- Un simulador de partidos.
- Un simulador completo de la fase de grupos.
- Un simulador completo de eliminatorias.
- Un sistema Monte Carlo para repetir el torneo miles de veces.

De esta forma se pueden obtener probabilidades de clasificación y de victoria final para cada selección.

# 2. Obtención de datos

Se han recopilado datos históricos de partidos internacionales junto con información actualizada de las selecciones participantes.

Fuentes utilizadas:

- FIFA Ranking
- Elo Rating
- Resultados históricos internacionales
- Rendimiento reciente de cada selección

Posteriormente se realizó un proceso de limpieza y transformación de los datos.

# 3. Ingeniería de variables

A partir de los datos originales se generaron variables predictoras para el modelo:

- Diferencia de ranking FIFA
- Diferencia de puntos FIFA
- Diferencia de Elo
- Forma reciente (últimos 5 partidos)
- Goles anotados recientemente
- Goles encajados recientemente
- Balance goleador reciente

Estas variables representan la diferencia de nivel entre dos selecciones antes de un partido.

# 4. Entrenamiento y comparación de modelos

Se entrenaron distintos algoritmos de Machine Learning para predecir la diferencia de goles esperada entre dos selecciones.

La variable objetivo utilizada fue:

Diferencia de goles = goles_local - goles_visitante

Modelos evaluados:

- Regresión Lineal
- Random Forest Regressor
- XGBoost Regressor
- Decision Tree Regressor
- Ridge Regression

Posteriormente se comparó el rendimiento de todos los modelos mediante métricas de evaluación como MAE, RMSE y R² para seleccionar el modelo más adecuado.

# 4.1. Selección del modelo final

Tras comparar los distintos algoritmos, la Regresión Lineal obtuvo el mejor equilibrio entre capacidad predictiva, estabilidad y velocidad de ejecución.

Por este motivo fue seleccionada como modelo principal para la simulación de partidos y del Mundial 2026.

# 5. Simulación de partidos

Una vez entrenado el modelo, se desarrolló una función capaz de simular partidos entre dos selecciones.

Proceso:

1. Construcción de variables del partido.
2. Escalado de características.
3. Predicción mediante el modelo.
4. Introducción de aleatoriedad mediante una distribución normal.
5. Conversión de la diferencia de goles en un marcador realista.

Esto permite obtener resultados distintos cada vez que se simula un partido.

# 6. Simulación de la fase de grupos

Se implementó un simulador completo de grupos.

Para cada grupo:

- Se generan todos los enfrentamientos.
- Se actualizan automáticamente las clasificaciones.
- Se calculan:
  - Partidos jugados
  - Victorias
  - Empates
  - Derrotas
  - Goles a favor
  - Goles en contra
  - Diferencia de goles
  - Puntos

Finalmente se obtiene la clasificación final de cada grupo.

# 7. Simulación de eliminatorias

Se implementó un sistema completo de eliminatorias:

- Dieciseisavos
- Octavos
- Cuartos
- Semifinales
- Final

En caso de empate se selecciona un ganador de forma aleatoria para representar prórroga o penaltis.

El sistema genera automáticamente el campeón del torneo.

# 8. Simulación completa del Mundial

Se desarrolló una función denominada:

simular_mundial()

Esta función:

1. Simula la fase de grupos.
2. Determina los clasificados.
3. Genera los cruces.
4. Simula todas las rondas eliminatorias.
5. Devuelve el campeón del torneo.

Con ello es posible reproducir un Mundial completo mediante una sola llamada.

# 9. Simulación Monte Carlo

Para obtener probabilidades reales se empleó el método Monte Carlo.

Consiste en repetir el Mundial cientos o miles de veces.

Ejemplo:

1000 simulaciones completas del torneo.

Cada simulación produce:

- Campeón
- Finalistas
- Semifinalistas
- Cuartofinalistas
- Clasificados a octavos

A partir de la frecuencia observada se calculan probabilidades.

# 10. Resultados preliminares

Tras 1000 simulaciones completas se obtuvieron los siguientes favoritos:

1. España
2. Argentina
3. Inglaterra
4. Francia
5. Países Bajos

Probabilidades estimadas de ser campeón:

España: 23.4%
Argentina: 21.8%

# 11. Optimización realizada

Mejoras implementadas:

- Eliminación de búsquedas repetidas en DataFrames.
- Simplificación de la generación de partidos.
- Reducción significativa de los tiempos de simulación.
